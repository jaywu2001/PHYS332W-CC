```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import struct
from scipy.signal import find_peaks as fp
```

```python
file = open("data/Feb10/50.bin","rb")
file.seek(0x10,0)
temp = file.read(8)
voltdiv = struct.unpack('<d',temp)[0]
file.seek(0x800,0)
x = np.fromfile(file,dtype=np.uint8)
x = np.int8(x-128)
x = (x)*(voltdiv/25)
n = fp(x,distance=len(x)/42)[0]
y = x[n]
n = np.linspace(0,len(y),len(y))
fig=plt.figure(figsize=(16,9),dpi=100)
plt.plot(np.linspace(0,len(y),len(x)),x)
plt.plot(n,y,'o')
plt.show()
```
