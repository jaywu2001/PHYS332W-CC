```python
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy.fft as fft
from scipy.signal import find_peaks as fp
import chaoticPackage as CC
import sim
```

```python
RV = np.array([])
DATA = np.array([])
r = np.arange(50,150.1,0.1)
for i in range(len(r)):
    temp = sim.simbifurcate(r[i]*1e3,0.2)
    RV = np.append(RV, np.ones(len(temp))*r[i])
    DATA = np.append(DATA, temp)
```

```python
CC.pack("simulated.bin",RV,DATA)
```
