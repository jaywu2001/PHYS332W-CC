```python
import matplotlib.pyplot as plt
import numpy as np
import os
import chaoticPackage as CC
import matplotlib.gridspec as gridspec
import sim as sim
```

```python
tfiles = ["SDS00051.bin","SDS00121.bin","SDS00261.bin","SDS00481.bin"]
tfiles = ["DATA/" + x for x in tfiles]

fig,plts = plt.subplots(nrows=2,ncols=2,tight_layout=1,height_ratios=(1,2))
fig.set_size_inches((8,6))
e1,e2,et = CC.getData(tfiles[1])
plts[0,0].plot(et,e1,'k')
plts[0,0].set_xlim([0,0.01])
plts[0,0].set_title("Experimental")
plts[0,0].set_xlabel(r"Time (s)")
plts[0,0].set_ylabel(r"Voltage (V)")

st,sx,sdx = sim.simulate(62e3,0.01)
plts[0,1].plot(st,sx,'k')
plts[0,1].set_xlim([800,1000])
plts[0,1].set_title("Simulated")
plts[0,1].set_xlabel(r"Time (s)")
plts[0,1].set_ylabel(r"Voltage (V)")

e1,e2,et = CC.getSmoothData(tfiles[1])
plts[1,0].plot(e1,-e2,'k')
plts[1,0].set_xlabel(r"$x$ (V)")
plts[1,0].set_ylabel(r"$\dot{x}$ (V)")

plts[1,1].plot(sx,sdx,'k')
plts[1,1].set_xlabel(r"$x$ (V)")
plts[1,1].set_ylabel(r"$\dot{x}$ (V)")
fig.suptitle(r"Experimental & Simulated Data at $R_v=62.0\text{k}\Omega$")
plt.show()

fig.savefig("wa2.pdf",format="pdf",bbox_inches="tight",transparent=1)
```
