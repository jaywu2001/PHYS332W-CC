```python
tfiles = ['SDS00001.bin','SDS00120.bin','SDS00211.bin','SDS00261.bin','SDS00481.bin']
tfiles = ["DATA/" + x for x in tfiles]
ch1data = []
ch2data = []
ch3data = []
colors = []


# # calculate color of dots based on max/min acceleration
# temp = (acceleration - acceleration.min())/(acceleration.max() - acceleration.min())
# # change values into RGBA array
# color = np.array([[1.0-x,x,0,1] for x in temp])
for i in range(5):
    t1,t2,t = CC.getSmoothData(tfiles[i])
    t3 = t2[1:]-t2[:-1]
    # t3 = t3[:50001]
    t /= 1e-7
    t1 = t1[:-1]
    t2 = t2[:-1]
    v2 = np.array(t3)/1e-7
    ch3data.append(v2[:50001])
    ch1data.append(t1[:50001])
    ch2data.append(t2[:50001])
    temp = v2    
    t = (temp - min(temp))/(max(temp) - min(temp))/2
    tc = np.array([[0.2+x,0.2+x,0.2+x,1] for x in t])
    colors.append(tc[:50001])
```

```python
plt.close("all")
plt.figure(figsize=(8,8),dpi=100)
plot = plt.axes(projection="3d")
x = 0
plot.plot3D(ch1data[x],-np.array(ch2data[x]),ch3data[x])
plot.plot3D(ch1data[x+1],-np.array(ch2data[x+1]),ch3data[x+1])
plot.plot3D(ch1data[x+2],-np.array(ch2data[x+2]),ch3data[x+2])
plot.plot3D(ch1data[x+3],-np.array(ch2data[x+3]),ch3data[x+3])
plot.plot3D(ch1data[4],-np.array(ch2data[4]),ch3data[4])
plt.show()
```
