```python
rv,d = CC.unpack("simulated.bin")
plt.figure(figsize=(8,4.5),dpi=100)
plt.plot(rv,d,'k.',markersize=1,alpha=0.7)
plt.xlabel(r"$R_v$ value (k$\Omega$)")
plt.ylabel(r"$x$ Amplitude (V)")
plt.title(r"$x$ Amplitude (V) vs. $R_v$ value (k$\Omega$) of a Simulated Chaotic System")
plt.xlim([50,150])
plt.show()
```
