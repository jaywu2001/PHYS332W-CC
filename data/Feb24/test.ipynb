```python
from parse import data as parse
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks as fp
import scipy.fft as fft
import bifurcation as BF
```

```python
datastrings = []
plt.figure(figsize=(8,4.5),dpi=150)
for i in range(80,100):
    string = "data/SDS000"+str(i)+".bin"
    datastrings.append(string)
for i in range(100,115):
    string = "data/SDS00"+str(i)+".bin"
    datastrings.append(string)
for i in range(0,len(datastrings)):
    data = BF.GETAMPLITUDES(datastrings[i])
    plt.plot(np.ones(data.size)*(50.0 + ((i+79)/10.0)), data, 'k.', markersize=1)
plt.ylabel(r"Amplitude (V)")
plt.xlabel(r"$R_v$ (k$\Omega$)")
plt.title(r"Amplitude (V) vs. $R_v$ (k$\Omega$)")
plt.show()
```
