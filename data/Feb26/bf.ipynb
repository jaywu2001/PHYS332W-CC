```python
import matplotlib.pyplot as plt
import numpy as np
import bifurcation as bf
import odecc as ODE
dt = 1/(1.4e6)
```

```python
STRINGS = []
for i in range(1,10):
    STRINGS.append("DATA/SDS0000"+str(i)+".bin")
for i in range(10,100):
    STRINGS.append("DATA/SDS000"+str(i)+".bin")
for i in range(100,1000):
    STRINGS.append("DATA/SDS00"+str(i)+".bin")
STRINGS.append("DATA/SDS01000.bin")
STRINGS.append("DATA/SDS01001.bin")
```

```python
plt.figure(figsize=(16,9),dpi=100)
for i in range(0,1001):
    DATA = bf.GETAMPLITUDES(STRINGS[i])
    DATA = DATA[np.argwhere(DATA < 2.3)]
    plt.plot(np.ones(DATA.size)*(50.0 + (i/10.0)), DATA, 'k.', markersize=0.1)
plt.ylabel(r"Amplitude (V)")
plt.xlabel(r"$R_v$ (k$\Omega$)")
plt.title(r"Amplitude (V) vs. $R_v$ (k$\Omega$)")
plt.show()
```

```python
plt.figure(figsize=(16,9),dpi=100)
for i in range(50,150):
    t,x,dx = ODE.sim(i*1e3,0.01)
    DATA = bf.SIMAMPLITUDES(x,t)
    plt.plot(np.ones(DATA.size)*(i), DATA, 'k.', markersize=1)
plt.show()
```
