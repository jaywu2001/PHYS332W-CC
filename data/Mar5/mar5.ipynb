```python
import matplotlib.pyplot as plt
import numpy as np
from parse import data as parse
import bifurcation as bf
import os
import chaoticPackage as CC
```

```python
files = os.listdir("alldata/")
```

```python
def strtorv(string):
    string = int(string[3:-4])
    return float(string)/100.0
```

```python
plt.figure(figsize=(16,9),dpi=150)
for i in files:
    string = "alldata/"+i
    _,_,data = CC.allData(string)
    res = np.ones(len(data))*strtorv(i)
    plt.plot(res,data,'ko',markersize=0.5,alpha=0.5)
plt.ylabel(r"Amplitude (V)")
plt.xlabel(r"$R_v$ (k$\Omega$)")
plt.title(r"Amplitude (V) vs. $R_v$ (k$\Omega$)")
plt.xticks(np.linspace(20.0,150.0,14))
plt.show()
```
