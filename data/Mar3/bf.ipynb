```python
import matplotlib.pyplot as plt
import numpy as np
import bifurcation as bf
import odecc as ODE
```

```python
STRINGS = []
# 25.0 - 49.9
for i in range(1002,1252):
    STRINGS.append("DATA/SDS0"+str(i)+".bin")
# 50 - 50.8
for i in range(1,10):
    STRINGS.append("../FEB26/DATA/SDS0000"+str(i)+".bin")
# 50.9 - 59.9
for i in range(10,100):
    STRINGS.append("../FEB26/DATA/SDS000"+str(i)+".bin")
# 60.0 - 150.0
for i in range(100,1000):
    STRINGS.append("../FEB26/DATA/SDS00"+str(i)+".bin")
STRINGS.append("../FEB26/DATA/SDS01000.bin")
STRINGS.append("../FEB26/DATA/SDS01001.bin")
```

```python
plt.figure(figsize=(16,9),dpi=100)
for i in range(0,len(STRINGS)):
    DATA = bf.GETAMPLITUDES(STRINGS[i])
    # DATA = DATA[np.argwhere(DATA < 2.3)]
    plt.plot(np.ones(DATA.size)*(25.0 + (i/10.0)), DATA, 'k.', markersize=0.1)
plt.ylabel(r"Amplitude (V)")
plt.xlabel(r"$R_v$ (k$\Omega$)")
plt.title(r"Amplitude (V) vs. $R_v$ (k$\Omega$)")
plt.show()
```
